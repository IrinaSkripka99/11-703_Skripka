```python
import plotly.graph_objects as go 
import numpy as np 
from sklearn.cluster import KMeans 
import math 
 
#colors = ("#FF0000", "#0000FF") 
# рандомно задаем точки 
n = 100  # количество точек 
# координаты точек для 3d 
x = np.random.randint(0, 100, n + 1) 
y = np.random.randint(0, 100, n + 1) 
z = np.random.randint(0, 100, n + 1) 
# должны каким-то образом разделить точки на классы 
# в частности применить k-means 
points = [] 
for i in range(n): 
    points.append([x[i], y[i], z[i]]) 
kmeans = KMeans(n_clusters=2, random_state=0).fit(points) 
clusters = kmeans.labels_ 
# print(clusters) 
colors = ['red']*n 
for i in range(n): 
    if clusters[i] == 1: 
        colors[i] = 'blue' 
# рисуем их при помощи plotly 
fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z, mode='markers', 
                                   marker=dict(color=colors))]) 
fig.show() 
```

```python
from sklearn.svm import SVC

svc = SVC(kernel='linear')
svc.fit(points[:n],clusters)

zz = lambda x,y: (-svc.intercept_[0]-svc.coef_[0][0]*x-svc.coef_[0][1]*y) / svc.coef_[0][2]

tmp = np.linspace(0,100,50)
xx,yx = np.meshgrid(tmp,tmp)

fig.add_trace(go.Surface(x=xx,y=yx,z=zz(xx,yx)))
fig.show()
```

```python
# предиктить новые точки (тоже на вход рандомно
x_new=np.random.randint(0,100)
y_new=np.random.randint(0,100)
z_new=np.random.randint(0,100)
points.append([x_new,y_new,z_new])

x[len(x)-1]=x_new
y[len(y)-1]=y_new
z[len(z)-1]=z_new
colors.append('green')
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def predict(data, weights):
    return sigmoid(np.dot(data, weights))
```

```python
def cluster(val):
    if(val >= 0.5):
        return 'blue'
    else:
        return 'red'
```

```python
def data_train(features, labels, weights, param, count):
    for i in range(count):
        weights = update_weights(features, labels, weights, param)
    return weights
```

```python
def update_weights(data, labels, weights, param):
    predictions = predict(data, weights)
    gradient = np.dot(np.transpose(data),  predictions - labels)
    gradient /= len(data)
    gradient *= param
    weights -= gradient
    return weights
```

```python
p = data_train(points[:(len(points)-1)], clusters, [0,0,0] , 0.001, 5000)
```
